```python
#Naming the columns in dataset 
item_col = ['movie id', 'movie title' ,'release date','video release date', 'IMDb URL', 'unknown', 'Action', 'Adventure',
    'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
    'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
item=pd.read_csv('ml-100k/u.item', sep='|',names=item_col, encoding='latin-1')
item
```

```python
#keeping only relavent columns in a dataset for further process
movie_genre=item[['movie id','movie title','Action', 'Adventure',
    'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
    'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']]
movie_genre
```

```python
#importing u.data dataset for the movie ratings
rating=pd.read_csv('ml-100k/u.data', sep='\t', encoding='latin-1')
```

```python
rating
```

```python
r_col=['user id','movie id','Rating','Timestamp']
rating=pd.read_csv('ml-100k/u.data', sep='\t',names=r_col, encoding='latin-1')
rating
```

```python
#merging two datasets into one for easy access.
movie_genre_rating = pd.merge(movie_genre,rating, on='movie id')
movie_genre_rating.head()
```

```python
#removing irrelavant column 'Timestamp'
movie_genre_rating.drop(['Timestamp'],axis=1,inplace=True)
movie_genre_rating
```

```python
#adding all the rating for a movie to see which has higher rating
movie_group=movie_genre_rating.groupby(['movie title','movie id','Action', 'Adventure',
    'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
    'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western'])['Rating'].sum()
```

```python
movieGroup=pd.DataFrame(movie_group)
movieGroup
```

```python
movieGroup.sort_values(by='Rating',ascending=False)
```

```python
#working on to find the top rated movies in each genre 
#but we could see duplicate entries 
moviesort=movie_genre_rating[movie_genre_rating['Rating']>3][['Action','movie title','Rating']]
moviesort
```

```python
moviesort.drop_duplicates(subset='movie title').sort_values('Rating', ascending = False).head(5)
```
